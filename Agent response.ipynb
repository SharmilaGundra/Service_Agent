```python
import os
import shutil
from pydub import AudioSegment
import librosa
import numpy as np
import speech_recognition as sr

# Set the directory containing your audio files
directory = r"C:\Users\gnr_n\Downloads\customers\New folder"
output_directory = r"C:\Users\gnr_n\Downloads\customers\labeled"

# Create the output directory if it doesn't exist
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

def extract_audio_features(file_path):
    y, sr = librosa.load(file_path)
    mfccs = np.mean(librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13).T, axis=0)
    pitch = np.mean(librosa.yin(y, fmin=50, fmax=300))  # Average pitch
    energy = np.mean(librosa.feature.rms(y=y))  # Average energy
    return np.hstack((mfccs, pitch, energy))

# Define thresholds for emotion detection
pitch_threshold = 150  # Adjust as needed
energy_threshold = 0.03  # Adjust as needed

# Initialize the speech recognizer
recognizer = sr.Recognizer()

# Loop through all files in the directory
for filename in os.listdir(directory):
    if filename.endswith(".mp3"):
        # Create full path for the mp3 file
        mp3_file_path = os.path.join(directory, filename)
        
        # Set the output wav file path
        wav_file_path = os.path.join(directory, f"{os.path.splitext(filename)[0]}.wav")
        
        # Convert the mp3 file to wav
        audio = AudioSegment.from_mp3(mp3_file_path)
        audio.export(wav_file_path, format="wav")
        
        print(f"Converted {mp3_file_path} to {wav_file_path}")
        
        # Extract features from the newly created wav file
        customer_features = extract_audio_features(wav_file_path)

        # Simulate emotion detection based on features
        emotion = "angry" if customer_features[-2] > pitch_threshold and customer_features[-1] > energy_threshold else "neutral"
        print(f"Detected emotion for {filename}: {emotion}")      
        
        # Save the labeled file if angry tone is detected
        if emotion == "angry":
            # Create a new filename
            new_file_name = f"angry_{os.path.basename(wav_file_path)}"
            new_file_path = os.path.join(output_directory, new_file_name)
            
            # Save the labeled file (copy the original file)
            shutil.copy(wav_file_path, new_file_path)
            print(f"Labeled audio saved as: {new_file_path}")

            # Convert the saved angry labeled file to text
            with sr.AudioFile(new_file_path) as source:
                audio_data = recognizer.record(source)  # Read the audio file
                try:
                    # Recognize the speech in the audio file
                    text = recognizer.recognize_google(audio_data)
                    print(f"Transcribed text for {new_file_name}: {text}")

                    # Save the transcribed text to a .txt file
                    text_file_path = os.path.join(output_directory, new_file_name.replace(".wav", ".txt"))
                    with open(text_file_path, "w") as text_file:
                        text_file.write(text)

                except sr.UnknownValueError:
                    print(f"Speech recognition could not understand audio for {new_file_name}.")
                except sr.RequestError as e:
                    print(f"Could not request results from Google Speech Recognition service for {new_file_name}; {e}")

        else:
            print("No angry tone detected. No file saved.")

print("All conversions, detections, and transcriptions completed.")
```
